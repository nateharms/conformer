```python
from autotst.conformer.ga import *
```

```python
smiles_dict = {
    "Phe" : "NC(CC1=CC=CC=C1)C(=O)O",
    "Gly" : "NCC(=O)O",
    "Ala" : "NC(C)C(=O)O",
    "Ile" : "NC(C(C)CC)C(=O)O",
    "Leu" : "NC(CC(C)C)C(=O)O",
    "Val" : "NC(C(C)C)C(=O)O",
    "Trp" : "NC(CC1=CNC2=C1(C=CC=C2))C(=O)O"
}
```

```python
smiles = smiles_dict.values()[0]
smiles = "CCCC"
```

```python
mol = AutoTST_Molecule(smiles)
```

```python
from hotbit.aseinterface import Hotbit
```

```python
mol.ase_molecule.set_calculator(Hotbit())
```

```python
mol.ase_molecule
mol.view_mol()
```

```python
from ase.optimize import BFGS
```

```python
opt = BFGS(mol.ase_molecule)
```

```python
opt.run(steps=10)
```

```python
mol.update_from_ase_mol()
mol.view_mol()
```

```python
results = perform_ga(mol)
```
