```python
mass = pd.DataFrame(mass)
import matplotlib.pyplot as plt
mass.columns = ["MM", "success", "mean", "std"]
fig, ax = plt.subplots()
#mass.sort_values("MM").plot.scatter( x="MM", y="success", marker="o",c="k", ax=ax, secondary_y=True)
mass.sort_values("MM").plot.scatter( x="MM", y="mean", yerr="std", marker="o",c="r", ax=ax )
```
