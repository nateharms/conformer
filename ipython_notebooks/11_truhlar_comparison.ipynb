```python
rxn = Multi_Reaction("OCCCC+[O]O_OCC[CH]C+OO", "H_Abstraction")
```

```python
rxn.multi_ts.view_ts()
```

```python
rxn.multi_ts.ase_ts.set_calculator(EMT())
rxn_initial_pop = create_initial_population(multi_object=rxn)
```

```python
perform_simple_es(rxn,
                  rxn_initial_pop,
                  top_percent=0.3,
                  tolerance=1e-4,
                  max_generations=100,
                  store_generations=True,
                  store_directory="./11_truhlar_comparison_files/")
```

```python
f = open("./11_truhlar_comparison_files/rxn_simple_es_generation_62.pkl", "r")
best = pickle.load(f)
best
```

```python
best.iloc[0][1:]

for dihedral in best.iloc[0][1:]:
    print dihedral
```

```python
for h, tor in enumerate(rxn.multi_ts.torsions):
    dihedral = best.iloc[0][h+1]
    i,j,k,l = tor.indices
    RHS = tor.right_mask
    rxn.multi_ts.ase_ts.set_dihedral(a1 = i,
                                    a2 = j, 
                                    a3 = k, 
                                    a4 = l, 
                                    angle= float(dihedral), 
                                    mask=RHS)
    
rxn.multi_ts.view_ts()
```
