# Setting up a reaction.

```python
rxn = Multi_Reaction("CCCCCC+[O]O_[CH2]CCCCC+OO", "H_Abstraction")
rxn.multi_ts.ase_ts.set_calculator(EMT())
rxn.multi_ts.view_ts()
```

## Taking all of the code from notebook 9 and adapting it for TSs

```python

TS = rxn.multi_ts

### Creating the initial population

possible_dihedrals = np.arange(0, 360+30, 30)

population_size = 100

population = []

for i in range(population_size):
    mol_copy = deepcopy(TS)
    dihedrals = []
    
    for torsion in mol_copy.torsions:
        dihedral = np.random.choice(possible_dihedrals)
        dihedrals.append(dihedral)
        i,j,k,l = torsion.indices
        RHS = torsion.RHS
        
        mol_copy.ase_ts.set_dihedral(i,j,k,l, dihedral)
    
    mol_copy.update_ts_from_ase_ts()
    
    
    e = mol_copy.ase_ts.get_potential_energy()
    
    population.append( [e] + dihedrals )
    
df = pd.DataFrame(population)
columns = ["Energy"]
for i in range(len(TS.torsion_list)):
    columns = columns +["Torsion " + str(i)] 
df.columns = columns
df = df.sort("Energy")

### Defining what is the top percent

top_percent = 0.30

top_population = population_size * top_percent

top = df.iloc[:int(top_population),:]


### Iterating through generations until this converges

q = 0
while (top.std()[0] / top.mean()[0] > 1e-4) or q < 150:
    
    print "This is the {}th generation".format(q)
    q = q + 1
    

    results = []
    for i in range(population_size):
        dihedrals = []
        for h, torsion in enumerate(TS.torsions):

            i,j,k,l = torsion.indices
            RHS = torsion.RHS

            dihedral = random.gauss(top.mean()[h+1], top.std()[h+1])
            dihedrals.append(dihedral)
            TS.ase_ts.set_dihedral(a1 = i,
                                    a2 = j, 
                                    a3 = k, 
                                    a4 = l, 
                                    angle= float(dihedral), 
                                    indices=RHS)

        e = TS.ase_ts.get_potential_energy()
        results.append([e] + dihedrals)

    df = pd.DataFrame(results)

    columns = ["Energy"]
    for i in range(len(TS.torsion_list)):
        columns = columns +["Torsion " + str(i)] 

    df.columns = columns
    df = df.sort("Energy")
    top = df.iloc[:int(top_population),:]

### Viewing the final results    
df       

```

## Performing the brute force method

```python
# Getting the torsion combos
torsion_list = TS.torsions
torsion_angles = np.arange(0, 360,30) ### You can change the degree step size
torsion_combos = list( itertools.combinations_with_replacement( torsion_angles, len(torsion_list)) )
if len(torsion_list) != 1:
    torsion_combos = list(
        set(
            torsion_combos + 
            list(itertools.combinations_with_replacement( 
                torsion_angles[::-1], len(torsion_list)
            ))))

    results = []
    
# Calculating the potential energy for each conformation
for combo in torsion_combos:
    geo = zip(torsion_list, combo)
    #print geo
    for torsion in geo:
        tor = torsion[0]
        #print tor
        angle = torsion[1]
        
        i,j,k,l = tor.indices
        RHS =  tor.RHS
        TS.ase_ts.set_dihedral(a1 = i,
                                a2 = j, 
                                a3 = k, 
                                a4 = l, 
                                angle= float(angle), 
                                indices=RHS)
    TS.update_ts_from_ase_ts()
    results.append([TS.ase_ts.get_potential_energy()] + list(combo) )

# Creating a dataframe of the results
brute_force = pd.DataFrame(results)
columns = ["Energy"]
for i in range(len(torsion_list)):
    columns = columns + ["Torsion " + str(i)] 

brute_force.columns = columns
brute_force = brute_force.sort("Energy")
brute_force
```
