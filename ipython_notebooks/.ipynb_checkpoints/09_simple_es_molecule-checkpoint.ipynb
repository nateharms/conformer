### Okay, we're going to be looking at conformer analysis of this molecule using ES

```python
mol = Multi_Molecule("COCCC")
mol.ase_molecule.set_calculator(EMT())
mol.view_mol()
```

### First, we're going to set up a population.

```python
possible_dihedrals = np.arange(0, 360+30, 30)

population_size = 200

population = []

for i in range(population_size):
    mol_copy = deepcopy(mol)
    dihedrals = []
    
    for torsion in mol_copy.torsions:
        dihedral = np.random.choice(possible_dihedrals)
        dihedrals.append(dihedral)
        i,j,k,l = torsion.indices
        RHS = torsion.RHS
        
        mol_copy.ase_molecule.set_dihedral(i,j,k,l, dihedral)
    
    mol_copy.update_geometry_from_ase_mol()
    
    
    e = mol_copy.ase_molecule.get_potential_energy()
    
    population.append( [e] + dihedrals )
    
df = pd.DataFrame(population)
columns = ["Energy"]
for i in range(len(mol.torsion_list)):
    columns = columns +["Torsion " + str(i)] 
df.columns = columns
df = df.sort("Energy")
df
```

### Next, we set what the top percent of our population is

```python
top_percent = 0.30

top_population = population_size * top_percent

top = df.iloc[:int(top_population),:]
```

### Then we are going to perform a single repopulation from the top percent parent population

```python
results = []
for i in range(population_size):
    dihedrals = []
    for h, torsion in enumerate(mol.torsions):
        
        i,j,k,l = torsion.indices
        RHS = torsion.RHS
        
        dihedral = random.gauss(top.mean()[h+1], top.std()[h+1])
        dihedrals.append(dihedral)
        mol.ase_molecule.set_dihedral(a1 = i,
                                a2 = j, 
                                a3 = k, 
                                a4 = l, 
                                angle= float(dihedral), 
                                indices=RHS)
        
    e = mol.ase_molecule.get_potential_energy()
    results.append([e] + dihedrals)
top = df.iloc[:int(top_population),:]
    
df = pd.DataFrame(results)

columns = ["Energy"]
for i in range(len(mol.torsion_list)):
    columns = columns +["Torsion " + str(i)] 
    
df.columns = columns
df = df.sort("Energy")
df
    
        
        
        
```

### Okay, now we're going to loop through this!

```python

q = 0
while (top.std()[1] > 0.01) and (top.std()[2] > 0.01):
    
    print "This is the {}th generation".format(q)
    q = q + 1
    

    results = []
    for i in range(population_size):
        dihedrals = []
        for h, torsion in enumerate(mol.torsions):

            i,j,k,l = torsion.indices
            RHS = torsion.RHS

            dihedral = random.gauss(top.mean()[h+1], top.std()[h+1])
            dihedrals.append(dihedral)
            mol.ase_molecule.set_dihedral(a1 = i,
                                    a2 = j, 
                                    a3 = k, 
                                    a4 = l, 
                                    angle= float(dihedral), 
                                    indices=RHS)

        e = mol.ase_molecule.get_potential_energy()
        results.append([e] + dihedrals)

    df = pd.DataFrame(results)

    columns = ["Energy"]
    for i in range(len(mol.torsion_list)):
        columns = columns +["Torsion " + str(i)] 

    df.columns = columns
    df = df.sort("Energy")
    top = df.iloc[:int(top_population),:]
df       
```

### ...and visualize the lowest energy conformer

```python


for h, tor in enumerate(mol.torsions):
    dihedral = df.iloc[0][h+1]
    i,j,k,l = tor.indices
    RHS = tor.RHS
    mol.ase_molecule.set_dihedral(a1 = i,
                                    a2 = j, 
                                    a3 = k, 
                                    a4 = l, 
                                    angle= float(dihedral), 
                                    indices=RHS)
    
print mol.torsions
mol.view_mol()
    
```

## Let's compare the results of the ES to the brute force method

```python

#mol.ase_molecule.set_calculator(calc)

# Getting the torsion combos
torsion_list = mol.torsions
torsion_angles = np.arange(0, 360,5) ### You can change the degree step size
torsion_combos = list( itertools.combinations_with_replacement( torsion_angles, len(torsion_list)) )
if len(torsion_list) != 1:
    torsion_combos = list(
        set(
            torsion_combos + 
            list(itertools.combinations_with_replacement( 
                torsion_angles[::-1], len(torsion_list)
            ))))

    results = []
    
# Calculating the potential energy for each conformation
for combo in torsion_combos:
    geo = zip(torsion_list, combo)
    #print geo
    for torsion in geo:
        tor = torsion[0]
        #print tor
        angle = torsion[1]
        
        i,j,k,l = tor.indices
        RHS =  tor.RHS
        mol.ase_molecule.set_dihedral(a1 = i,
                                a2 = j, 
                                a3 = k, 
                                a4 = l, 
                                angle= float(angle), 
                                indices=RHS)
    mol.update_geometry_from_ase_mol()
    results.append([mol.ase_molecule.get_potential_energy()] + list(combo) )

# Creating a dataframe of the results
brute_force = pd.DataFrame(results)
columns = ["Energy"]
for i in range(len(torsion_list)):
    columns = columns + ["Torsion " + str(i)] 

brute_force.columns = columns
brute_force = brute_force.sort("Energy")
brute_force
```

### The torsions are similar! So let's visualize the lowest energy conformer for the brute force method

```python

tor0, tor1 = mol.torsions

for i, tor in enumerate(mol.torsions):
    dihedral = brute_force.iloc[0][i+1]
    i,j,k,l = tor.indices
    RHS = tor.RHS
    mol.ase_molecule.set_dihedral(a1 = i,
                                    a2 = j, 
                                    a3 = k, 
                                    a4 = l, 
                                    angle= float(dihedral), 
                                    indices=RHS)
mol.view_mol()
    
```

### Okay, let's try this for a more complex molecule

For this we will be looking at `CC(CC)CCCC`

```python
mol = Multi_Molecule("CC(CC)CCCC") #Choosing one of the biofules hehe
mol.ase_molecule.set_calculator(EMT())


### Creating the initial population

possible_dihedrals = np.arange(0, 360+30, 30)

population_size = 200

population = []

for i in range(population_size):
    mol_copy = deepcopy(mol)
    dihedrals = []
    
    for torsion in mol_copy.torsions:
        dihedral = np.random.choice(possible_dihedrals)
        dihedrals.append(dihedral)
        i,j,k,l = torsion.indices
        RHS = torsion.RHS
        
        mol_copy.ase_molecule.set_dihedral(i,j,k,l, dihedral)
    
    mol_copy.update_geometry_from_ase_mol()
    
    
    e = mol_copy.ase_molecule.get_potential_energy()
    
    population.append( [e] + dihedrals )
    
df = pd.DataFrame(population)
columns = ["Energy"]
for i in range(len(mol.torsion_list)):
    columns = columns +["Torsion " + str(i)] 
df.columns = columns
df = df.sort("Energy")

### Defining what is the top percent

top_percent = 0.30

top_population = population_size * top_percent

top = df.iloc[:int(top_population),:]


### Iterating through generations until this converges

q = 0
while (top.std()[0] / top.mean()[0] > 1e-5):
    
    print "This is the {}th generation".format(q)
    q = q + 1
    

    results = []
    for i in range(population_size):
        dihedrals = []
        for h, torsion in enumerate(mol.torsions):

            i,j,k,l = torsion.indices
            RHS = torsion.RHS

            dihedral = random.gauss(top.mean()[h+1], top.std()[h+1])
            dihedrals.append(dihedral)
            mol.ase_molecule.set_dihedral(a1 = i,
                                    a2 = j, 
                                    a3 = k, 
                                    a4 = l, 
                                    angle= float(dihedral), 
                                    indices=RHS)

        e = mol.ase_molecule.get_potential_energy()
        results.append([e] + dihedrals)

    df = pd.DataFrame(results)

    columns = ["Energy"]
    for i in range(len(mol.torsion_list)):
        columns = columns +["Torsion " + str(i)] 

    df.columns = columns
    df = df.sort("Energy")
    top = df.iloc[:int(top_population),:]

### Viewing the final results    
df       

```
