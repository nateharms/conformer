```python
from autotst.conformer.ga import *
```

```python
mol = AutoTST_Molecule("NCC(=O)O")
mol.view_mol()
```

```python
min_rms = np.arange(10,130, 10)
min_rms
```

```python
from hotbit.aseinterface import Hotbit
mol.ase_molecule.set_calculator(Hotbit())
```

```python
results = []
for m in min_rms:
    _, confs = perform_ga(autotst_object=mol, min_rms=m)
    results.append([m, len(confs)])
    print "For a RMS of {}, we identified {} unique conformers.".format(m, len(confs))
    
```
